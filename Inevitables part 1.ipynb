```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
os.chdir('D:/Data sets')
```

```python
df= pd.read_csv('new store.csv')
```

```python
df
```

```python
plt.scatter(df['Customer ID'],df['Profit'])
plt.axhline(0,color="r",linestyle='-')
```

```python
X = df["Customer ID"]
y = df["Profit"]
```

```python
import statsmodels.api as sm
model1 = sm.OLS(y,X).fit()
predictions = model1.predict(X)
model1.summary()
```

```python
# Split data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.1, random_state=1)
```

```python
X_train
```

```python
y_train
```

```python
X_train=X_train.values.reshape((-1,1))
```

```python
X_train
```

```python
from sklearn import linear_model as lm
model=lm.LinearRegression()
results=model.fit(X_train,y_train)
```

```python
results
```

```python
#score function
accuracy=model.score(X_train,y_train)
print("Accuracy of the model:",accuracy)
```

```python
#print the coff
print("intercept:",model.intercept_)
print("slope",model.coef_)
```

```python
X_test
```

```python
X_test=X_test.reshape((-1,1))
```

```python
X_test
```

```python
#predictions for the model
predictions=model.predict(X_test)
print("predicted profits",predictions,sep="\n")
```

```python
plt.scatter(y_test,predictions)
```

```python
#make a prediction
Xnew=[[20365]]
#make a prediction
ynew=model.predict(Xnew)
```

```python
ynew
```

```python
#Evaluating the model 
from sklearn.metrics import mean_squared_error, r2_score
X_train = X_train.reshape(-1,1)
y_train_prediction = model.predict(X_train)

X_test = X_test.reshape(-1,1)
y_test_prediction = model.predict(X_test)
```

```python
# printing values
print('Slope:' ,model.coef_)
print('Intercept:', model.intercept_)
print("\n")

# model evaluation for training set
import numpy as np
rmse_training = (np.sqrt(mean_squared_error(y_train, y_train_prediction)))
r2_training = r2_score(y_train, y_train_prediction)

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse_training))
print('R2 score is {}'.format(r2_training))
print("\n")

# model evaluation for testing set
rmse_testing = (np.sqrt(mean_squared_error(y_test, y_test_prediction)))
r2_testing = r2_score(y_test, y_test_prediction)

print("The model performance for testing set")
print("--------------------------------------")
print('Root mean squared error: ', rmse_testing)
print('R2 score: ', r2_testing)
```
